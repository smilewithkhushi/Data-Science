```python
import pandas as pd
df = pd.DataFrame({'NAME':['Aditya','Ken','Ram'],
                   'AGE':[18,20,19],
                   'DEPARMENT':['CS','Maths','Chemistry'],
                   'MONTH':['September','July','February'],
                   'CGPA':[9.5,9,7.8]})
data={"NAME" : "Niya", "AGE":21, "DEPARTMENT":"Biology", "MONTH":"March","CGPA":9.5}
print("Dataframe: \n",df)
df=df.append(data,ignore_index=True)   #adding new row to dataframe
print(df)         
hobby=["Chess","Writing","Yoga","Cricket"]
df=df.assign(HOBBY=hobby)         #adding new column to dataframe
print(df)
df=df.drop(["MONTH"], axis=1)     #deleting  the column "MONTH"
print(df)
df=df.drop(0)                     #delete the row with index 0
print(df)
```
